```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
회사정보 =pd.read_csv('./data/train_job/job_companies.csv')
잡키워드 = pd.read_csv('./data/train_job/job_tags.csv')
태그키워드 = pd.read_csv('./data/train_job/tags.csv')
train = pd.read_csv('./data/train_job/train.csv')
유저태그 = pd.read_csv('./data/train_job/user_tags.csv')
테스트 = pd.read_csv('./data/test_job.csv')
```

```python
회사키워드 = pd.merge(회사정보,공지키워드, on = 'jobID', how='left')
del 회사키워드['jobID']
del 회사키워드['companySize']
```

```python
castings = [(list(set(x['companyID'].values))[0], x['keyword'].values) for index, x in 회사키워드.groupby('companyID')]
회사키워드2 = pd.DataFrame(data=castings, columns=['companyID','keyword'])
회사키워드2 = 회사키워드2.set_index('companyID')
회사키워드2.head()
```

```python
회사키워드2
```

```python
회사정보2 = 회사정보[['companyID','companySize']]
회사키워드3 = pd.merge(회사키워드2, 회사정보2 , on = 'companyID', how='left')
회사키워드3.drop_duplicates(['companyID'], inplace=True)
```

```python
회사정보그룹 = 회사정보.groupby('companyID').count()
del 회사정보그룹['companySize']
```

```python
회사키워드3 = 회사키워드3.merge(회사정보그룹, on = 'companyID', how='left')
```

```python
회사키워드3.sort_values(by='jobID', ascending=False).tail(15)
```

```python
회사키워드3.to_csv('./companykeyword.csv')
```
